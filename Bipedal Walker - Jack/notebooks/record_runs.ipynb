```python
import gymnasium as gym
import numpy as np
import pandas as pd
```

```python
run_data = pd.read_csv("notebooks/runs/bipedal_walker/run_2.csv")
run_data_actions = run_data["actions"].values
run_data_actions
```

```python
actions = []
for action in run_data_actions:
    actions.extend(eval(action))
actions = np.array(actions)
```

```python
env = gym.make("BipedalWalker-v3", render_mode="human")
obs, _ = env.reset()
```

```python
for action in actions:
    obs, _, _, _, _ = env.step(action)
    env.render()
```
