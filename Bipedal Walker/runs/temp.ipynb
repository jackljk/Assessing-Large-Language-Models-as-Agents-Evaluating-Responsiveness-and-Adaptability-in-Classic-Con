```python
import pandas as pd
```

```python
baseline_4o = pd.read_csv("baseline/gpt-4o/observations_0.csv")
baseline_4o_mini = pd.read_csv("baseline/gpt-4o-mini/observations_1.csv")
enhanced_4o = pd.read_csv("enhanced/gpt-4o/observations_0.csv")
enhanced_4o_mini = pd.read_csv("enhanced/gpt-4o-mini/observations_1.csv")
```

```python
def count_steps(df):
    # lenght of df * 5
    return len(df) * 5

def get_total_score(df):
    return df["score"].iloc[-1]
```

```python
# Get total steps for each 
runs = [baseline_4o, baseline_4o_mini, enhanced_4o, enhanced_4o_mini]
names = ["baseline_4o", "baseline_4o_mini", "enhanced_4o", "enhanced_4o_mini"]
for i, run in enumerate(runs):
    print(f"Total steps for {names[i]}: {count_steps(run)}")
```

```python
# Get total score for each
for i, run in enumerate(runs):
    print(f"Total score for {names[i]}: {get_total_score(run)}")
```
